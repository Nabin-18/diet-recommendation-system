```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("recipes.csv")
```

```python
# finding missing values
missing_values = dataset.isnull().sum()
```

```python
missing_values_in_percentage=missing_values/dataset.shape[0] * 100
```

```python
# finding duplicates value 
duplicates = dataset.duplicated().sum()
```

```python
# identifying garbage values 
for i in dataset.select_dtypes(include="object").columns:
  print(dataset[i].value_counts())
  print("--------------------------------------------------")
   
```

```python
# //remove garbage values
dataset = dataset.dropna()
dataset = dataset.drop_duplicates()

```

```python
# // EXPOLATORY DATA ANALYSIS
#descriptive statistics
dataset.describe().T
```

```python
#descriptive statistics for categorical data
dataset.describe(include="object").T
```

```python
#histogram to visualize the distribution of numerical data
dataset.hist(figsize=(15, 10), bins=20)
plt.tight_layout()
plt.show()
```

```python
#box plot to identify outliers
plt.figure(figsize=(15, 10))
sns.boxplot(data=dataset, orient="h")
plt.tight_layout()
```

```python
#scatter plot to visualize the relationship between two numerical variables
plt.figure(figsize=(15, 10))
sns.scatterplot(data=dataset, x="Calories", y="FatContent")
plt.title("Scatter plot of calories vs total fat")
plt.xlabel("Calories")
plt.ylabel("Total Fat")
plt.show()
```

```python
#correlation with heatmap to interpret the relation and multicollinearity

dataset.select_dtypes(include="number").corr()
```
